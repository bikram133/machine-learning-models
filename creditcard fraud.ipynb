```python
import numpy as np                    
import pandas as pd      
```

```python
df1=pd.read_csv('creditcard.csv')
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
df1['Amount'] = scaler.fit_transform(df1['Amount'].values.reshape(-1,1))
```

```python
df1.head()
```

```python
df1.info()
```

```python

df1['Amount_log'] = np.log(df1['Amount'] + 1)  
```

```python
df1.columns
```

```python
df1['Class'].value_counts()
```

```python
print(df1[df1['Class']==1])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(C=0.5, max_iter=1000, class_weight='balanced')
x=df1.iloc[:, df1.columns != 'Class']
y=df1['Class']
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
model.fit(x_train, y_train)
predicted=model.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predicted))
print(classification_report(y_test, predicted))

```

```python
print(model.score(x_test, y_test))
```
