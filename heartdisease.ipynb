```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('heartdisease.csv')
df1.head()
```

```python
df1['num_binary'] = df1['num'].apply(lambda x: 0 if x == 0 else 1)
```

```python
df1.head()
```

```python
df1['num'].value_counts()
```

```python
df1.drop(columns=['num'], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_oldpeak= scaler.fit_transform(df1[['oldpeak']])
df1['oldpeak'] = scaled_oldpeak
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
transformed = encoder.fit_transform(df1[['cp', 'thal', 'slope']])
```

```python
df1.info()
```

```python
df1.drop(columns=['ca','thal'],inplace=True)
```

```python
df1.info()
```

```python
df1.fillna(value=df1[['trestbps','chol','thalch','oldpeak']].mean(), inplace=True)
df1.fillna(value=df1[['restecg','slope']].mode().iloc[0], inplace=True)
```

```python
df1.info()
```

```python

df1.replace({'True':1, 'False':0}, inplace=True)
```

```python
df1.fillna(value=df1[['fbs','exang']].mode().iloc[0], inplace=True)
```

```python
df1.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
df1[["cp"]] = encoder.fit_transform(df1[["cp"]])
```

```python
df1.info()
```

```python
df1.corr
```

```python
df1.head()
```

```python
df1.drop('cp','restecg','slope', axis=1, inplace=True)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(learning_rate=0.1, n_estimators=100, max_depth=3)
from sklearn.model_selection import train_test_split
x=df1[['id','age','sex','trestbps','chol','fbs','thalch','exang','oldpeak']]
y=df1['num_binary']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
score=model.score(x_test, y_test)
print("Model accuracy:", score)

```
